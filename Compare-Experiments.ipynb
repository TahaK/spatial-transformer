```python
import matplotlib.pyplot as plt
import pickle
import pandas as pd
import numpy as np
import os
```

```python
metrics_1_dir = "metrics-learning_rate-0.0001-with_pooling"
metrics_2_dir = "metrics-learning_rate-0.0001-without_pooling"
figures_dir = "figure-compare-learning_rate-0.001-with-without-pooling"
```

```python
if not os.path.exists(figures_dir):
    os.makedirs(figures_dir)
```

```python
metrics_list = []
for i in range(0,10):
    i_metrics = np.array(list(map(lambda x: x[0], pickle.load( open( metrics_1_dir+"/metrics"+str(i)+".pickle", "rb" )))))
    metrics_list.append(i_metrics)
metrics_with_pooling = np.stack(metrics_list)
```

```python
metrics_list = []
for i in range(0,10):
    i_metrics = np.array(list(map(lambda x: x[0], pickle.load( open( metrics_2_dir+"/metrics"+str(i)+".pickle", "rb" )))))
    metrics_list.append(i_metrics)
metrics_without_pooling = np.stack(metrics_list)
```

```python
datasets = ['normal','imbalanced','augmented']
batch_sizes =['10','20','50','100','256']
training_set_percentages =['1', '10', '40' ,'80', '100']
```

```python
indexes = []
for id_ in range(len(datasets)):
        for training_set_percentage in training_set_percentages:
            for batch_size in batch_sizes:
                indexes.append((datasets[id_],training_set_percentage,batch_size))
```

```python
list(enumerate(indexes))
```

```python
plt.figure(figsize=(30,10))

plt.subplot(1,5,1)
plt.xlabel('Configuration')
plt.ylabel('Accuracy')
plt.ylim([0.1,1.])
plt.title('Batch size 256 %100 of the data')
plt.boxplot(np.hstack((metrics_with_pooling[:,[24]],metrics_without_pooling[:,[24]])))
plt.xticks([1, 2],['Pooling','No internal Pooling'])


plt.subplot(1,5,2)
plt.xlabel('Configuration')
plt.ylabel('Accuracy')
plt.ylim([0.1,1.])
plt.title('Batch size 256 %80 of the data')
plt.boxplot(np.hstack((metrics_with_pooling[:,[19]],metrics_without_pooling[:,[19]])))
plt.xticks([1, 2],['Pooling','No internal Pooling'])


plt.subplot(1,5,3)
plt.xlabel('Configuration')
plt.ylabel('Accuracy')
plt.ylim([0.1,1.])
plt.title('Batch size 256 %40 of the data')
plt.boxplot(np.hstack((metrics_with_pooling[:,[14]],metrics_without_pooling[:,[14]])))
plt.xticks([1, 2],['Pooling','No internal Pooling'])


plt.subplot(1,5,4)
plt.xlabel('Configuration')
plt.ylabel('Accuracy')
plt.ylim([0.1,1.])
plt.title('Batch size 256 %10 of the data')
plt.boxplot(np.hstack((metrics_with_pooling[:,[9]],metrics_without_pooling[:,[9]])))
plt.xticks([1, 2],['Pooling','No internal Pooling'])


plt.subplot(1,5,5)
plt.xlabel('Configuration')
plt.ylabel('Accuracy')
plt.ylim([0.1,1.])
plt.title('Batch size 256 %1 of the data')
plt.boxplot(np.hstack((metrics_with_pooling[:,[4]],metrics_without_pooling[:,[4]])))
plt.xticks([1, 2],['Pooling','No internal Pooling'])


plt.savefig(figures_dir+'/experiments_comparison-Normal-Batch256.png')
plt.show()
```
